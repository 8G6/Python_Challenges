```python
#3) Write a program that counts the number of words, vowels, consonants and
#   special characters in a string. ( special characters @ # $ % )
#   Go to the site https://miniwebtool.com/random-quote-generator/ Generate a random quote . Give
#  it as an input. Give the quote and the result as a comment 

#KNP19EC010
#Script By Devadut S Balan

x=input("Enter a string : ")

#A simple function for checkiing if the nourn is sigular or plural
def check_s(c):
    if c>1:
        return "s"
    else:
        return ''
    
#Function for checking the given character is capital case
def check_cap(c):
    if ord(c)>64 and ord(c)<90:
        return True
    else:
        return False

#Function for checking the given character is small case    
def check_small(c):
    if ord(c)>96 and ord(c)<122:
        return True
    else:
        return False

#Main Function
def myfun(a):
    #Spliting the input in to word into an array and getting the length of array
    word_num=len(a.split(' '))
    
    #Making the input as a list of character
    a=list(a)
    
    #Defing all needed things
    caps_vow=["A","E","I","O","U"]
    small_vow=['a','e','i','o','u']
    char=['@','#','$','%']
    
    #Initializing all varables
    caps_vow_num,small_vow_num,caps_cons_num,small_cons_num,char_num=0,0,0,0,0
    
    #An array for storing the singlar or plural case
    s=[]
    
    #Checking each character by character
    for i in a:
        if i in caps_vow:
            caps_vow_num+=1
        elif check_cap(i):
            caps_cons_num+=1
        elif i in small_vow:
            small_vow_num+=1
        elif check_small(i):
            small_cons_num+=1
        elif i in char:
            char_num+=1
    t=[word_num,caps_vow_num,caps_cons_num,small_vow_num,small_cons_num,char_num]
        
    #Cheking  the singlar or plural case
    for i in t:
        s.append(check_s(i))
    #Joining the array back to orginal input
    
    a="".join(a)
    
    #Output
    print('The given string '+a+' have :')
    print('                             1) '+str(word_num)+' word'+s[0])
    print('                             2) '+str(caps_vow_num)+' Upper Case Vowel'+s[1])
    print('                             3) '+str(caps_cons_num)+' Upper Case Consonant'+s[2])
    print('                             4) '+str(small_vow_num)+' Lower Case Vowel'+s[3])
    print('                             5) '+str(small_cons_num)+' Lower Case Consonant'+s[4])
    print('                             6) '+str(char_num)+' Special Character'+s[5])
myfun(x)        

#input :I will age ungracefully until I become an old woman in a small garden, doing whatever the hell I want %$.
#output

#                             1) 21 words
#                             2) 3 Upper Case Vowels
#                             3) 0 Upper Case Consonant
#                             4) 29 Lower Case Vowels
#                             5) 49 Lower Case Consonants
#                             6) 2 Special Characters

#I have added 2 special Characters at the end to show the working of Special Characters in my script
```
