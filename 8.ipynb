```python
#8) Write a program to convert numbers between the octal and hexadecimal representations of integers.


#KNP19EC010
#Script By Devadut S Balan

#Some colors for better apperrance 
red='\033[31m'
green='\033[32m'
rst = '\033[0m'
#Decimal to Octal using inbuit function
def octal_inbuit(a):
    return oct(a)[2:]

#Decimal to Hexadecimal using inbuit function
def hex_inbuit(a):
    return hex(a)[2:]

#Decimal to Octal in the old way
def octal(a):
    s=[]
    while a>0:
        #storing the reminder when diving with 8
        s.append(a%8)
        a=a//8
    #reversing convering to string and joining as a string converting it to int 
    return int("".join([str(i) for i in s[::-1]]))

#Decimal to Hexadecimal in the old way
def hexadec(a):
    s=[]
    while a>0:
        s.append(a%16)
        a=a//16
    for i in range(len(s)):
        if s[i]>9:
            s[i]=chr(s[i]+55)
        else:
            s[i]=str(s[i])
    return "".join(s[::-1])

#Octal to decimal
def octal_to_decimal(a):
    a=list(str(a))
    dec=0
    a=a[::-1]
    #converting octal to decimal
    for i in range(len(a)):
        dec+=(int(a[i])*(8**i))
    return dec

#Hexadecimal to decimal
def hex_to_decimal(a):
    a=list(a)
    dec=0
    a=a[::-1]
    C=['A','B','C','D','E','F']
    c=['a,','b','c','d','f']
    #converting octal to decimal
    for i in range(len(a)):
        if a[i] in C:
            a[i]=ord(a[i])-55
        elif a[i] in c:
            a[i]=ord(a[i])-87
    for i in range(len(a)):
        dec+=(int(a[i])*(16**i))
    return dec

#Octal to hexadecimal convertion
def octal_to_hex(a):
    return hexadec(octal_to_decimal(a))

#Hexadecimal to octal convertion
def hex_to_octal(a):
    return octal(hex_to_decimal(a))
while True:
    print(rst,"\n")
    x=input('Select any convertion \n             1.Decimal to Octal\n             2.Decimal to Hexadecimal\n             3.Octal to Decimal\n             4.Hexadecimal to Decimal\n             5.Octal to Hexadecimal\n             6.Hexadecimal to Octal\n             q.To exit\nSelect any digit (1-6 or q) for convertion : ')
    print('\n\n')
    if x=='q':
        print(red,"You have exied the program")
        break
    elif int(x)>0 and int(x)<7:
        x=int(x)
    else:
        print(red,"Entered a wrong input retry")
    print(rst)
    if x==1:
        print("You have selected Decimal to Octal\n")
        p=int(input("Enter a Decimal number to convert : "))
        print(green,p," is ",octal(p)," in Octal")
    if x==2:
        print("You have selected Decimal to Hexadecimal\n")
        p=int(input("Enter a Decimal number to convert : "))
        print(green,p," is ",hexadec(p)," in Hexadecimal")
    if x==3:
        print("You have selected Octal to Decimal\n")
        p=int(input("Enter a Octal number to convert : "))
        print(green,p," is ",octal_to_decimal(p)," in Decimal")
    if x==4:
        print("You have selected Hexadecimal to Decimal\n")
        p=input("Enter a Hexadecimal number to convert : ")
        print(green,p," is ",hex_to_decimal(p)," in Decimal")
    if x==5:
        print("You have selected Octal to Hexadecimal\n")
        p=int(input("Enter a Octal number to convert : "))
        print(green,p," is ",octal_to_hex(p)," in Hexadecimal")
    if x==6:
        print("You have selected Hexadecimal to Octal\n")
        p=int(input("Enter a Hexadecimal number to convert : "))
        print(green,p," is ",hex_to_octal(p)," in Octal")
```
